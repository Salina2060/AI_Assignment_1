**1. Flatten a Nested List:**
Write a Python function that takes a nested list and flattens it into a single list, where all the elements
are in a single dimension.

**• Requirements:**
– Define a function flatten(lst) that takes a nested list lst and returns a flattened version
of the list.– Example: For the input [[1, 2], [3, 4], [5]], the output should be [1, 2, 3, 4, 5]

```python
def flatten(lst):
    """
    Flattens a list of lists into a single list.
    """
    flat_list = []
    for sublist in lst:
        for item in sublist:
            flat_list.append(item)
    return flat_list

flatten([[1, 2], [3, 4], [5]])
```

**3. Sum of All Elements in a Nested List:**
Write a Python function that calculates the sum of all the numbers in a nested list (regardless of depth).

**• Requirements**:
– Define a function sum
nested(lst) that takes a nested list lst and returns the sum of all
the elements.– Example: For the input [[1, 2], [3, [4, 5]], 6], the output should be 21

```python
def sum_nested(lst):
    total = 0
    # 'To-Do List' starts with the main list
    todo_list = [lst]

    while todo_list:
        #taking the last item out of our 'To-Do List'
        current_item = todo_list.pop()

        for item in current_item:
            if isinstance(item, list):
                #if we find another list, add it to the 'To-Do List'
                todo_list.append(item)
            else:
                # If it's a number, add it to the total
                total += item

    return total


print(sum_nested([[1, 2], [3, [4, 5]], 6]))
```

**6. Count Occurrences of an Element in a Nested List:**
Write a Python function that counts how many times a specific element appears in a nested list.

**• Requirements:**– Define a function count
occurrences(lst, elem) that counts the occurrences of elem in
the nested list lst.– Example: Fortheinput lst = [[1, 2], [2, 3], [2, 4]]andelem = 2,theoutputshould
be 3

```python
def count_occurrences(lst, elem):
    count = 0
    todo_list = [lst]

    while todo_list:
        current_item = todo_list.pop()

        for item in current_item:
            if isinstance(item, list):
                # If it's a list, save it for later
                todo_list.append(item)
            elif item == elem:
                # If it matches our number, count it
                count += 1

    return count

print(count_occurrences([[1, 2], [2, 3], [2, 4]], 2))
```

**7. Flatten a List of Lists of Lists:**
Write a Python function that flattens a list of lists of lists into a single list, regardless of the depth.

**• Requirements:**– Define a function deep
flatten(lst) that takes a deeply nested list lst and returns a single
f
lattened list.– Example: For the input [[[1, 2], [3, 4]], [[5, 6], [7, 8]]], the output should be
[1, 2, 3, 4, 5, 6, 7, 8]

```python
def deep_flatten(lst):
    result = []
    for item in lst:
        if isinstance(item, list):
            # use .extend to add items from the inner list to our result
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result

deep_flatten([[[1, 2]], [[3, 4]]])
```

• Using NumPy arrays, perform matrix multiplication of two matrices of size 1000x1000.
Measure and print the time taken for this operation.

```python
import numpy as np
import time

size_mat = 1000

# using NumPy Arrays
#creating 1000x1000 matrices filled with ones
mat_a_np = np.ones((size_mat, size_mat))
mat_b_np = np.ones((size_mat, size_mat))

start_np = time.time()
#standard matrix multiplication in NumPy
res_np = np.dot(mat_a_np, mat_b_np)
end_np = time.time()

print(f"NumPy Matrix Multiplication (Size 1000): {end_np - start_np:.5f} sec")


#using Python Lists
#Creating 1000x1000 lists filled with ones
list_a = [[1.0] * size_mat for _ in range(size_mat)]
list_b = [[1.0] * size_mat for _ in range(size_mat)]
#initialising the result list with zeros
res_list = [[0.0] * size_mat for _ in range(size_mat)]

start_list = time.time()
# Using 3 nested loops for Matrix Multiplication
for i in range(size_mat):
    for j in range(size_mat):
        for k in range(size_mat):
            res_list[i][j] += list_a[i][k] * list_b[k][j]
end_list = time.time()

print(f"Python List Matrix Multiplication (Size 1000): {end_list - start_list:.5f} sec")
```
