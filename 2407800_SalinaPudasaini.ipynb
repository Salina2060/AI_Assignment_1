**Task - 3 - Outlier Detection:**

**Tasks:**

1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
2. Define outliers using the 1.5 Ã— IQR rule.
3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color.
4. Discuss the characteristics of these outliers and their potential impact on regional averages

```python
#Calculation for Outliers
#CALCULATION OF Q1, Q3 and IQR FOR SCORE
Q1_score = south_asia_df['score'].quantile(0.25)
Q3_score = south_asia_df['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

#CALCULATION OF Q1, Q3 and IQR FOR GDP per capita
Q1_GDP = south_asia_df['Log GDP per capita'].quantile(0.25)
Q3_GDP = south_asia_df['Log GDP per capita'].quantile(0.75)
IQR_GDP = Q3_GDP - Q1_GDP

#CALCULATION OF LOWER BOUND AND UPPER BOUNDS
lower_bound_score = Q1_score - 1.5 *  IQR_score
upper_bound_score = Q3_score + 1.5*  IQR_score

lower_bound_GDP = Q1_GDP - 1.5 *  IQR_GDP
upper_bound_GDP = Q3_GDP + 1.5 *  IQR_GDP

#IDENTIFICATION OF OUTLIERS
outliers_score = south_asia_df[(south_asia_df['score'] < lower_bound_score) | (south_asia_df['score'] > upper_bound_score)]
outliers_GDP = south_asia_df[(south_asia_df['Log GDP per capita'] < lower_bound_GDP) | (south_asia_df['Log GDP per capita'] > upper_bound_GDP)]

print(outliers_score['Country name'])
print(outliers_GDP['Country name'])
```

Here is the Scatter Plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers in a different color:

```python
#plotting the normal or reqular points
plt.scatter(south_asia_df['Log GDP per capita'],south_asia_df['score'],color='green',label='Normal')
#plotting the outliers
plt.scatter(outliers_GDP['Log GDP per capita'],outliers_score['score'],color='blue',label='Outliers')
plt.title ('Scatter plot with GDP per Capita and Score')
plt.xlabel('GDP per Capita')
plt.ylabel('Score')
# plt.xlim(0,)
# plt.ylim(0,)
plt.legend()
plt.grid()
plt.show ()
```
